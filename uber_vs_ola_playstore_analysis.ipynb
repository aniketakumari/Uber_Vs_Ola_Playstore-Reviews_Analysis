# **Data Visualization**

## **1**. **Rating Distribution**

```python
plt.figure(figsize=(10, 6))

# Ola Ratings Distribution
plt.subplot(1, 2, 1)
sns.countplot(x='rating', data=ola, hue='rating', palette="plasma", legend=False)
plt.title('Ola Ratings Distribution')

# Uber Ratings Distribution
plt.subplot(1, 2, 2)
sns.countplot(x='rating', data=uber, hue='rating', palette="plasma", legend=False)
plt.title('Uber Ratings Distribution')

plt.tight_layout()
plt.show()
```

## **Conclusion** : Ola has got most 5 star voting then uber.

## **2**. **Sentiment Analysis**

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(str(text))
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

ola['sentiment'] = ola['review_description'].apply(get_sentiment)
uber['sentiment'] = uber['review_description'].apply(get_sentiment)

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.countplot(x='sentiment', hue='sentiment', palette="pastel", data=ola)
plt.title('Ola Sentiment Analysis')

plt.subplot(1, 2, 2)
sns.countplot(x='sentiment', hue='sentiment', palette="pastel", data=uber)
plt.title('Uber Sentiment Analysis')

plt.tight_layout()
plt.show()
```

## **Conclusion** : Ola has got most positive feedback than uber from people.

## **3**. **Developer Response Analysis**

```python
ola['sentiment'] = ola['developer_response'].apply(get_sentiment)
uber['sentiment'] = uber['developer_response'].apply(get_sentiment)

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.countplot(x='sentiment',hue='sentiment', palette="pastel", data=ola)
plt.title('Ola Sentiment Analysis')

plt.subplot(1, 2, 2)
sns.countplot(x='sentiment',hue='sentiment', palette="pastel", data=uber)
plt.title('Uber Sentiment Analysis')

plt.tight_layout()
plt.show()
```

## **Conclusion** : Ola has got most positive feedback than uber from developers.

## **4**. **Pie chart distribution of sources for both Ola & Uber.**

```python
# Count occurrences of each source
ola_source_counts = ola['source'].value_counts()
uber_source_counts = uber['source'].value_counts()

# Create pie charts
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.pie(ola_source_counts, labels=ola_source_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sources - Ola')

plt.subplot(1, 2, 2)
plt.pie(uber_source_counts, labels=uber_source_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sources - Uber')

plt.tight_layout()
plt.show()
```

## **Conclusion** : Distribution of sources for both Ola & Uber is Google Play Store.

## **5**. **Time series analysis on the review posted for the Apps Ola & Uber.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert review_date to datetime format, coercing errors
ola['review_date'] = pd.to_datetime(ola['review_date'], errors='coerce')
uber['review_date'] = pd.to_datetime(uber['review_date'], errors='coerce')

# Drop rows with invalid dates
ola.dropna(subset=['review_date'], inplace=True)
uber.dropna(subset=['review_date'], inplace=True)

# Set review_date as the index
ola.set_index('review_date', inplace=True)
uber.set_index('review_date', inplace=True)

# Resample data to get a daily count of reviews
ola_daily_review_counts = ola.resample('D').size()
uber_daily_review_counts = uber.resample('D').size()

# Plot time series of daily review counts for Ola
plt.figure(figsize=(10, 6))
ola_daily_review_counts.plot(marker='o', color='blue')
plt.title('Daily Review Counts - Ola')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()

# Plot time series of daily review counts for Uber
plt.figure(figsize=(10, 6))
uber_daily_review_counts.plot(marker='o', color='green')
plt.title('Daily Review Counts - Uber')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```

## **Conclusion** : Ola has got most number of review counts than uber.
